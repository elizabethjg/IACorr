# Download and load catalogue

## Query to get the catalogue from CosmoHUB

```sql
SELECT 
    x_gal, y_gal, z_gal, 
    `ra_gal`, `dec_gal`, 
    `true_redshift_gal`, 
    `eps1_gal`, `eps2_gal`, 
    (-2.5 * log10(sdss_g_el_model3_ext_odonnell_ext) - 48.6) as mag_sdss_g, 
    (-2.5 * log10(sdss_r_el_model3_ext_odonnell_ext) - 48.6) as mag_sdss_r, 
    (-2.5 * log10(sdss_i_el_model3_ext_odonnell_ext) - 48.6) as mag_sdss_i, 
    (-2.5 * log10(sdss_r_abs) - 48.6) as mag_abs_sdss_r 
FROM fs2_mock_std_2_1_10b_phz_wise 
WHERE
    (-2.5 * log10(sdss_r_abs) - 48.6) < -22. 
  AND 
    true_redshift_gal < 0.36 
  AND 
    true_redshift_gal > 0.16
```

# Compute correlations

## Set-up parameters

````python
# parameters related with bins
nbins = 10 # number of radial bins
rmin = 0.1 # minimum value for rp (r in case of the quadrupole)
rmax = 50. # maximum value for rp (r in case of the quadrupole)
pi_max = 30. # maximum value along l.o.s. (Pi) 
npi = 5 # number of bins in Pi
mubins = 10 # number of bins in mu

# Related to JK patches
NPatches = int(nbins**(3./2.))
print('Number of patches',NPatches)
# Other configuration parameters
ncores = 30 # Number of cores to run in parallel
slop = 0. # Resolution for treecorr
box = False # Indicates if the data corresponds to a box, otherwise it will assume a lightcone
exact_position = True # Indicates if the coordinates are exactly provided (e.g. simulated data 
# without any error added in the position). Otherwise it will assume that the positions are not
# exact and will use ra, dec to match the catalogues. If this parameter is set as True, it
# will neglect box and it will set it as False
sky_threshold = 1.0 # Threshold for matching the catalogues in arcsecond, used if exact_position
#is set to True.
grid_resolution = 10 # Resolution of the grid used to compute the quadrupole, a lower number will
#speed up the computation but reaches to larger radial distances.
````

## Make configuration input

````python
config = {'col_names': col_names,
          'nbins': nbins,
          'rmin': rmin,
          'rmax': rmax,
          'pi_max': pi_max,
          'npi': npi,
          'grid_resolution': grid_resolution,
          'mubins': mubins,
          'NPatches': NPatches,
          'ncores': ncores,
          'slop': slop,
          'box': box,
          'exact_position': exact_position,
          'sky_threshold': sky_threshold,
         }
          
````

````python
q = np.array(cat['Projected_Axis_Lengths']).T[0]/(np.array(cat['Projected_Axis_Lengths'])).T[1]
````

````python
'Number of objects = '+str(np.sum(m))
````

## Make input catalogue

````python
col_names_box = ['x_gal','y_gal','z_gal','eps1_gal','eps2_gal']
````

````python
x = np.array(cat['Semimajor_Axis_Direction']).T[0][m]
y = np.array(cat['Semimajor_Axis_Direction']).T[1][m]

e = ((1 - q ) / (1 + q ))  # size of ellipticity
R = 1 - np.mean(e ** 2) / 2.0  # responsivity factor ???????
e /= (2.*R)  # size of ellipticity
theta = np.arctan2(y,x)
````

````python
df[g1] = -1*e*np.cos(2*theta)
df[g2] = -1*e*np.sin(2*theta)
````
